[Amazon] Given an integer K and a queue of integers, we need to reverse the order of the first K elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

- enqueue(x) : Add an item x to rear of queue

- dequeue() : Remove an item from front of queue

- size() : Returns number of elements in queue.

- front() : Finds front item.

**Note**: The above operations represent the general processings. In-built functions of the respective languages can be used to solve the problem.

**If k is bigger than queue size you must return "queue size is smaller than k"**

Expected Time Complexity : O(n)


```python
class Queue :
  def __init__(self, data) :
    self._data = data
    self._size = len(data)
  
  def size(self) :
    return self._size

  def front(self) :
    return self._data[0]
  
  def dequeue(self) :
    if self._size==0 :
      return "Empty queue"
    else :
      self._size-=1
      return self._data.pop(0)
    
  def enqueue(self,x):
    self._size+=1
    self._data.append(x)
```

```python
def modify_queue(queue,k) :
  """
  >>> modify_queue(Queue([1,2,3,4,5,5]),3)
  [3, 2, 1, 4, 5, 5]
  >>> modify_queue(Queue([]),7)
  'queue size is smaller than k'
  >>> modify_queue(Queue([1,2,3,4,5]),3)
  [3, 2, 1, 4, 5]
  >>> modify_queue(Queue([1,2,3,4]),4)
  [4, 3, 2, 1]
  >>> modify_queue(Queue([1,2,3,4]),5)
  'queue size is smaller than k'
  >>> modify_queue(Queue([10,2,3,4,5,2,3,4]),3)
  [3, 2, 10, 4, 5, 2, 3, 4]
  >>> modify_queue(Queue([10,2,3,4,5,2,3,4]),1)
  [10, 2, 3, 4, 5, 2, 3, 4]
  """
  index = 0
  size = queue.size()

  if size < k:
    return 'queue size is smaller than k'
  for item in queue:
    if item != k:
      index += 1
    else:
      return rev_queue(Queue[:index+1])+Queue[index+1:]
  return queue

def rev_queue(prev_queue) : 
  new_queue = Queue()
  for item in prev_queue:
    prev_queue.dequeue()
    new_queue.enqueue(item)
  new_queue.reverse()
  return new_queue
```

[Accolite / Amazon] Given an non-negative integer x, find the square root of x. If x is not a perfect square, then return floor(√x).

<br/>

**Note:**

Expected Time Complexity: O(log N)


```python
def floor_sqrt(x):
    '''
    >>> floor_sqrt(0)
    0
    >>> floor_sqrt(1)
    1
    >>> floor_sqrt(3)
    1
    >>> floor_sqrt(4)
    2
    >>> floor_sqrt(5)
    2
    >>> floor_sqrt(38808)
    196
    >>> floor_sqrt(38809)
    197
    >>> floor_sqrt(38810)
    197
    >>> floor_sqrt(39204)
    198
    '''
    low = 0
    high = x

    while low <= high:
      mid = low + (high - low) // 2

      if mid ** 2 > x:
        high = mid - 1
      elif mid ** 2 < x:
        low = mid + 1 
      else:
        return mid

    return high ##왜 high를 return? 
```
